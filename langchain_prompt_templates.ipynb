## Task 1: Replace Hardcoded Prompt with PromptTemplate

**Goal:** Convert a plain Python function that returns a hardcoded string into a proper LangChain `PromptTemplate`.

**Before (hardcoded):**
```python
def explain_topic(topic):
    return "Explain " + topic + " in simple terms for beginners"
```

**After:** Use `PromptTemplate` with a `{topic}` input variable.

````python
# ============================================================
# TASK 1: Replace Hardcoded Prompt with PromptTemplate
# ============================================================

# Define the template string using a placeholder variable
# No f-strings used — LangChain handles variable injection
EXPLAIN_TEMPLATE = "Explain {topic} in simple terms for beginners"

# Create the PromptTemplate object
explain_prompt = PromptTemplate(
    input_variables=["topic"],
    template=EXPLAIN_TEMPLATE
)

# Reusable function that uses the PromptTemplate (no f-strings)
def explain_topic(topic):
    """
    Generate a beginner-friendly explanation prompt for a given topic.
    Uses PromptTemplate instead of a hardcoded string.

    Args:
        topic (str): The subject to explain.

    Returns:
        str: The fully formatted prompt string.
    """
    # .format() injects the variable into the template safely
    return explain_prompt.format(topic=topic)


# ---- Test Task 1 ----
print("=" * 55)
print("TASK 1 OUTPUT")
print("=" * 55)

test_topics = ["Machine Learning", "Blockchain", "Quantum Computing"]

for t in test_topics:
    result = explain_topic(t)
    print("Topic     :", t)
    print("Prompt    :", result)
    print("-" * 50)
````

## Task 2: Multi-Input Prompt System

**Goal:** Build a `PromptTemplate` that accepts three variables: `topic`, `audience`, and `tone`.

Test it with three different input combinations.

````python
# ============================================================
# TASK 2: Multi-Input Prompt System
# ============================================================

# Template with three input variables
MULTI_INPUT_TEMPLATE = "Explain {topic} for {audience} in a {tone} tone"

# Create the PromptTemplate with all three input variables declared
multi_input_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=MULTI_INPUT_TEMPLATE
)


def generate_multi_input_prompt(topic, audience, tone):
    """
    Generate a customised explanation prompt using three inputs.

    Args:
        topic    (str): Subject to explain.
        audience (str): Target audience (e.g., beginners, kids, engineers).
        tone     (str): Desired tone (e.g., friendly, fun, technical).

    Returns:
        str: The formatted prompt string.
    """
    return multi_input_prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    )


# ---- Test Task 2 with three combinations ----
print("=" * 55)
print("TASK 2 OUTPUT")
print("=" * 55)

# Each entry is a (topic, audience, tone) tuple
test_combinations = [
    ("AI",            "beginners", "friendly"),
    ("Python",        "kids",      "fun"),
    ("Deep Learning", "engineers", "technical"),
]

for topic, audience, tone in test_combinations:
    prompt_output = generate_multi_input_prompt(topic, audience, tone)
    print("Inputs : topic={}, audience={}, tone={}".format(topic, audience, tone))
    print("Prompt :", prompt_output)
    print("-" * 50)
````

## Task 3: Prompt Variations Engine

**Goal:** Create three distinct `PromptTemplate` objects for different styles:
- **Teaching** — Step-by-step explanation
- **Interview** — Ask 3 questions
- **Storytelling** — Explain as a story

Run all three for `topic = "Machine Learning"`.

````python
# ============================================================
# TASK 3: Prompt Variations Engine
# ============================================================

# --- Template definitions (no f-strings, no hardcoded topic) ---

TEACHING_TEMPLATE = "Explain {topic} step by step"

INTERVIEW_TEMPLATE = "Ask 3 interview questions about {topic}"

STORYTELLING_TEMPLATE = "Explain {topic} as an engaging story"

# --- Create PromptTemplate objects for each style ---

teaching_prompt = PromptTemplate(
    input_variables=["topic"],
    template=TEACHING_TEMPLATE
)

interview_prompt = PromptTemplate(
    input_variables=["topic"],
    template=INTERVIEW_TEMPLATE
)

storytelling_prompt = PromptTemplate(
    input_variables=["topic"],
    template=STORYTELLING_TEMPLATE
)

# --- Store all variants in a dict for easy lookup ---
PROMPT_VARIANTS = {
    "teaching":     teaching_prompt,
    "interview":    interview_prompt,
    "storytelling": storytelling_prompt,
}


def run_prompt_variations(topic):
    """
    Run all three prompt style variations for a given topic.

    Args:
        topic (str): The subject to use across all variations.

    Returns:
        dict: A mapping of style name to formatted prompt string.
    """
    results = {}
    for style_name, template in PROMPT_VARIANTS.items():
        results[style_name] = template.format(topic=topic)
    return results


# ---- Test Task 3 ----
print("=" * 55)
print("TASK 3 OUTPUT")
print("=" * 55)

variations = run_prompt_variations("Machine Learning")

for style, prompt_text in variations.items():
    print("Style  :", style.upper())
    print("Prompt :", prompt_text)
    print("-" * 50)
````

## Task 4: ChatPromptTemplate System

**Goal:** Use `ChatPromptTemplate` with two message roles:
- **system** — Sets the AI's behavioural role (teacher / interviewer / motivator)
- **human** — Provides the topic input

This mirrors how real chat models (GPT-4, Claude, etc.) receive structured conversation history.

````python
# ============================================================
# TASK 4: ChatPromptTemplate System
# ============================================================

# --- System message template: defines the AI's role ---
# {role} is injected at runtime — no hardcoding
SYSTEM_TEMPLATE = "You are a helpful {role}. Always be clear, concise, and helpful."

# --- Human message template: the user's actual request ---
HUMAN_TEMPLATE = "Please teach me about {topic}."

# --- Build the ChatPromptTemplate from message tuples ---
# Format: list of (role_string, template_string) tuples
chat_prompt = ChatPromptTemplate.from_messages([
    ("system", SYSTEM_TEMPLATE),
    ("human",  HUMAN_TEMPLATE),
])


def build_chat_prompt(role, topic):
    """
    Build a structured chat prompt with a system role and human message.

    Args:
        role  (str): The AI's role (e.g., teacher, interviewer, motivator).
        topic (str): The subject the user wants to learn about.

    Returns:
        list: A list of formatted message objects (SystemMessage, HumanMessage).
    """
    # .format_messages() returns a list of typed message objects
    messages = chat_prompt.format_messages(role=role, topic=topic)
    return messages


def display_chat_messages(messages):
    """
    Pretty-print a list of chat message objects.

    Args:
        messages (list): List of LangChain message objects.
    """
    for msg in messages:
        # Use the class name to identify the role cleanly
        role_label = type(msg).__name__.replace("Message", "").upper()
        print("[{}]  {}".format(role_label, msg.content))


# ---- Test Task 4 with multiple role/topic pairs ----
print("=" * 55)
print("TASK 4 OUTPUT")
print("=" * 55)

chat_test_cases = [
    ("teacher",     "Neural Networks"),
    ("interviewer", "Data Structures"),
    ("motivator",   "Python Programming"),
]

for role, topic in chat_test_cases:
    print("\nRole: {} | Topic: {}".format(role, topic))
    print("-" * 50)
    msgs = build_chat_prompt(role, topic)
    display_chat_messages(msgs)
    print()
````

## Task 5: Input Validation Layer

**Goal:** Write a `validate_inputs(audience, tone)` function that enforces allowed values.

- **audience** must be one of: `beginner`, `intermediate`, `expert`
- **tone** must be one of: `formal`, `casual`, `fun`
- If invalid, raise a `ValueError` with a descriptive message OR assign safe default values.

````python
# ============================================================
# TASK 5: Input Validation Layer
# ============================================================

# Define allowed values as constants (easy to update in one place)
VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
VALID_TONES     = ["formal", "casual", "fun"]

# Safe default values used when strict_mode is False
DEFAULT_AUDIENCE = "beginner"
DEFAULT_TONE     = "casual"


def validate_inputs(audience, tone, strict_mode=True):
    """
    Validate audience and tone inputs against allowed values.

    Args:
        audience    (str):  The target audience string.
        tone        (str):  The desired tone string.
        strict_mode (bool): If True, raise ValueError on invalid input.
                            If False, assign default values silently.

    Returns:
        tuple: (validated_audience, validated_tone)

    Raises:
        ValueError: When strict_mode=True and an invalid value is provided.
    """
    # Normalise to lowercase to make validation case-insensitive
    audience = audience.strip().lower()
    tone     = tone.strip().lower()

    # --- Validate audience ---
    if audience not in VALID_AUDIENCES:
        if strict_mode:
            raise ValueError(
                "Invalid audience '{}'. Allowed values: {}".format(
                    audience, VALID_AUDIENCES
                )
            )
        else:
            print("[Warning] Invalid audience '{}'. Defaulting to '{}'".format(
                audience, DEFAULT_AUDIENCE
            ))
            audience = DEFAULT_AUDIENCE

    # --- Validate tone ---
    if tone not in VALID_TONES:
        if strict_mode:
            raise ValueError(
                "Invalid tone '{}'. Allowed values: {}".format(
                    tone, VALID_TONES
                )
            )
        else:
            print("[Warning] Invalid tone '{}'. Defaulting to '{}'".format(
                tone, DEFAULT_TONE
            ))
            tone = DEFAULT_TONE

    return audience, tone


# ---- Test Task 5 ----
print("=" * 55)
print("TASK 5 OUTPUT")
print("=" * 55)

# Test 1: Valid inputs
print("\nTest 1 — Valid inputs:")
a, t = validate_inputs("beginner", "fun")
print("Result: audience={}, tone={}".format(a, t))

# Test 2: Invalid inputs, default fallback (strict_mode=False)
print("\nTest 2 — Invalid inputs with default fallback:")
a, t = validate_inputs("newbie", "aggressive", strict_mode=False)
print("Result: audience={}, tone={}".format(a, t))

# Test 3: Invalid inputs, strict mode raises error
print("\nTest 3 — Invalid inputs with strict mode (ValueError expected):")
try:
    validate_inputs("newbie", "aggressive", strict_mode=True)
except ValueError as e:
    print("Caught ValueError:", e)
````

## Task 6: Prompt Generator App

**Goal:** Build a `generate_prompt(topic, audience, tone, style)` function that:
1. Validates inputs
2. Selects the correct template based on `style`
3. Returns the final, formatted prompt string

Supported styles: `teaching`, `interview`, `storytelling`

````python
# ============================================================
# TASK 6: Prompt Generator App
# ============================================================

# --- Style-specific templates that include audience + tone context ---
# Each template uses all four inputs for full personalisation

GENERATOR_TEMPLATES = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=(
            "Explain {topic} step by step for {audience} in a {tone} teaching style"
        )
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=(
            "Ask 3 {tone} interview questions about {topic} "
            "suitable for a {audience}-level candidate"
        )
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=(
            "Explain {topic} for {audience} in a {tone} storytelling style"
        )
    ),
}

VALID_STYLES = list(GENERATOR_TEMPLATES.keys())


def generate_prompt(topic, audience, tone, style):
    """
    Full prompt generator with validation and style selection.

    Args:
        topic    (str): The subject to explain.
        audience (str): Target audience — beginner / intermediate / expert.
        tone     (str): Desired tone — formal / casual / fun.
        style    (str): Prompt style — teaching / interview / storytelling.

    Returns:
        str: The final validated and formatted prompt.

    Raises:
        ValueError: If style is not one of the supported options.
    """
    # Step 1: Validate audience and tone (strict mode)
    audience, tone = validate_inputs(audience, tone, strict_mode=True)

    # Step 2: Validate style separately
    style = style.strip().lower()
    if style not in VALID_STYLES:
        raise ValueError(
            "Invalid style '{}'. Allowed values: {}".format(style, VALID_STYLES)
        )

    # Step 3: Select the correct template and format it
    selected_template = GENERATOR_TEMPLATES[style]
    final_prompt = selected_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

    return final_prompt


# ---- Test Task 6 ----
print("=" * 55)
print("TASK 6 OUTPUT")
print("=" * 55)

test_requests = [
    ("Neural Networks",  "beginner",    "fun",      "storytelling"),
    ("Data Structures",  "intermediate","casual",   "teaching"),
    ("System Design",    "expert",      "formal",   "interview"),
]

for topic, audience, tone, style in test_requests:
    output = generate_prompt(topic, audience, tone, style)
    print("Inputs : topic={}, audience={}, tone={}, style={}".format(
        topic, audience, tone, style
    ))
    print("Output :", output)
    print("-" * 50)
````

## Task 7: Template Reusability Test

**Goal:** Demonstrate that a **single** `PromptTemplate` can produce meaningfully different outputs when given different inputs.

This proves that templates are reusable, dynamic, and not hardcoded.

````python
# ============================================================
# TASK 7: Template Reusability Test
# ============================================================

# ONE shared template used for all 5 test inputs below
REUSABLE_TEMPLATE = "Explain {topic} to a {audience} in a {tone} way"

reusable_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=REUSABLE_TEMPLATE
)


def test_reusability(inputs_list):
    """
    Run one template against multiple input sets to demonstrate reusability.

    Args:
        inputs_list (list of dicts): Each dict must have keys: topic, audience, tone.

    Returns:
        list of str: Formatted prompt strings.
    """
    results = []
    for inputs in inputs_list:
        formatted = reusable_prompt.format(
            topic=inputs["topic"],
            audience=inputs["audience"],
            tone=inputs["tone"]
        )
        results.append(formatted)
    return results


# ---- 5 different input combinations for the same template ----
five_inputs = [
    {"topic": "Transformers",          "audience": "beginner",     "tone": "simple"},
    {"topic": "Recursion",             "audience": "kids",         "tone": "fun"},
    {"topic": "Cloud Computing",       "audience": "manager",      "tone": "formal"},
    {"topic": "Reinforcement Learning","audience": "researcher",   "tone": "technical"},
    {"topic": "APIs",                  "audience": "entrepreneur",  "tone": "casual"},
]

print("=" * 55)
print("TASK 7 OUTPUT — Template Reusability Test")
print("=" * 55)
print("Template used:", REUSABLE_TEMPLATE)
print("-" * 55)

reuse_results = test_reusability(five_inputs)

for idx, result in enumerate(reuse_results, start=1):
    print("Test {} : {}".format(idx, result))
````

## Final Pipeline: User Input → Validation → PromptTemplate → Output

This cell ties everything together into one clean, end-to-end pipeline function.

**Flow:**
```
User Input
    ↓
validate_inputs()       ← Task 5
    ↓
generate_prompt()       ← Task 6 (selects template, formats output)
    ↓
build_chat_prompt()     ← Task 4 (wraps in chat role structure)
    ↓
Final Prompt Output
```

````python
# ============================================================
# FINAL PIPELINE
# User Input → Validation → PromptTemplate → Chat Output
# ============================================================

def full_prompt_pipeline(topic, audience, tone, style, role="teacher"):
    """
    Complete end-to-end pipeline: takes raw user inputs and produces
    a fully structured, validated chat prompt.

    Args:
        topic    (str): Subject to explain.
        audience (str): Target audience.
        tone     (str): Tone of explanation.
        style    (str): Style — teaching / interview / storytelling.
        role     (str): System role for ChatPromptTemplate.

    Returns:
        dict: Contains 'plain_prompt' and 'chat_messages' keys.
    """
    print("[Step 1] Raw Inputs Received")
    print("        topic={}, audience={}, tone={}, style={}, role={}".format(
        topic, audience, tone, style, role
    ))

    # Step 2: Validate audience and tone
    print("[Step 2] Validating inputs...")
    try:
        audience, tone = validate_inputs(audience, tone, strict_mode=False)
        print("        Validated: audience={}, tone={}".format(audience, tone))
    except ValueError as ve:
        print("        Validation error:", ve)
        return None

    # Step 3: Generate the plain prompt using style template
    print("[Step 3] Generating plain prompt via PromptTemplate...")
    try:
        plain_prompt = generate_prompt(topic, audience, tone, style)
        print("        Plain Prompt:", plain_prompt)
    except ValueError as ve:
        print("        Style error:", ve)
        return None

    # Step 4: Wrap into a chat structure using ChatPromptTemplate
    print("[Step 4] Building ChatPromptTemplate with role='{}'...".format(role))
    chat_messages = build_chat_prompt(role=role, topic=plain_prompt)

    print("[Step 5] Final Chat Prompt Messages:")
    display_chat_messages(chat_messages)

    return {
        "plain_prompt":  plain_prompt,
        "chat_messages": chat_messages,
    }


# ---- Run the full pipeline ----
print("=" * 55)
print("FINAL PIPELINE OUTPUT")
print("=" * 55)

pipeline_output = full_prompt_pipeline(
    topic    = "Neural Networks",
    audience = "beginner",
    tone     = "fun",
    style    = "storytelling",
    role     = "teacher"
)

print()
print("Pipeline completed successfully!")
print("Plain Prompt stored:", pipeline_output["plain_prompt"])
````
